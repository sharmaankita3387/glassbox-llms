# Chain of Thought (CoT) Reasoning

## Overview of CoT
Chain of thought (CoT) reasoning refers to the way an LLM reasons about problems. Specifically, when an LLM is prompted to think through its answer, it usually produces its logic with a series of steps. For example, if prompted to add 5 and 3, an AI model may say: start with 5, then count up 3. This series of steps, or chain of thought, is critical in improving AI performance as well as an important area of research for mechanistic interpretability. 

## How to measure the faithfulness of CoT reasoning?
CoT is an incredibly powerful tool, yet is that really what is going on within the model? This is an important question to ask, as understanding the underlying structure of these LLMs can not only allow us to improve them further, but also to develop safety mechanisms. Due to their nature as black boxes, it is a challenge to truly understand the inner mechanisms of any given AI model, yet there are ways to test whether their reasoning is actually how they think about a problem, or if its just a distraction so that they fulfill the goals of the prompt they are given. The way we can measure if they are actually using CoT is to test for something called post-hoc reasoning. Post-hoc means done after the event, and as such, post-hoc reasoning is reasoning done by a model after it already has the answer. The more post-hoc the answer, the less faithful it is, as the reasoning is done after the fact, and we cant check if that is the same process the model used or if its just making a reasonable guess about what a logical chain of reasoning might be. The question then becomes, how do we test for post-hoc reasoning?

## How to test for post-hoc reasoning?
There are a variety of methods of testing for post-hoc reasoning. I recommend reading Lanham et. al,  2023 for deeper insight into these testing methods. 
One way of testing post-hoc reasoning is to test if the answer given to some question changes based upon how much of the reasoning is provided to the model. 



This figure shows the percentage of reasoning provided to a model and the relationship to the percentage of the answer, which is the same as the full CoT. The more consistency between low and high percent of reasoning provided indicates high amounts of post-hoc reasoning, as the CoT doesnt markedly affect the reasoning. Each line on this graph is a different set of questions that the AI has to answer. Furthermore, this graph shows the difference between different step amounts, how many steps the AI takes to come to a conclusion. 
Another method is to add mistakes during the CoT and see how that affects the models answer.



What is being done here is that the mistake is placed along some percentage of the CoT, so the later the mistake, the easier it should be to ignore. In this case, if a models answer changes when its CoT changes, if its not post-hoc, then the answer should change as well. We see similar trends here, the same sets of questions following similar patterns. This shows us that certain types of questions generate more post-hoc reasoning than others. 
There are many other ways to test for post-hoc reasoning, but these are the most straightforwardly understandable and provide a good knowledge base for what kinds of tests can be run to test for post-hoc reasoning. Furthermore, there are many insights in these graphs which provide questions such as, why are some types of questioning more post-hoc than others?

## Is CoT reasoning faithful?
From the previous examples, we can infer that CoT isnt a one-stop solution for faithful reasoning, given the discrepancy between the amount of post-hoc reasoning in different question sets. This discrepancy is notable because it means CoT is only faithful some of the time, and is therefore not a reliable test for faithfulness in all cases. This problem is exacerbated further with larger models. 



In order to be a reliable indicator of faithful reasoning, the answer with and without CoT should have low similarity. It seems, though, that when model size increases, CoT becomes less reliable. In fact, 6 of the 8 question sets have the most faithful responses at ~10 billion parameters. This kind of trend is worrying, as we create larger and larger models, we are less able to measure their faithfulness accurately with CoT. 
Furthermore, studies suggest that the apparent reasoning prowess of Chain-of-Thought (CoT) is largely a brittle mirage (Zhao et. al, 2025). Their reasoning for this is that, when tested on problems outside of their dataset, LLMs will generate plausible but flawed reasoning chains. They argue that this is more dangerous than a simple wrong answer, as it seems dependable, given its human-like reasoning, when, really, it is unreliable.

## Can CoT tests be relied upon for mechanistic interpretability?
It depends not only on the size of the model being tested, but also on the questions being asked. CoT is an active field of research, and we still have yet to know everything about how reliable it is. There are two cases: CoT is reliable some of the time, or it is almost always unreliable. Overall, CoT, at this time, is too unreliable to be a core test for interpretability because of the inconsistency and prevalence of post-hoc reasoning across different tasks and model sizes. That being said, there are many cases where CoT can reveal interesting insights into the internal mechanisms of an LLM. For example, say an LLM is queried with some question, using percent reasoning, adding mistakes, and other CoT tests can provide useful data for other modules. CoT can be leveraged to create more robust testing scenarios and serve as a critical auxiliary tool by providing access to deeper analysis of the model's internal states during the sequential generation process.
